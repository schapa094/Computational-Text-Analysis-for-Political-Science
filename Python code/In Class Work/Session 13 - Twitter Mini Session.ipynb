```python
import nltk
tweet1=tweets_text[0]
tokenized_tweet1 = nltk.word_tokenize(tweet1)
tokenized_tweet1
```

```python
without_capital_letters = [word.lower() for word in tokenized_tweet1]
without_capital_letters
```

```python
from nltk.corpus import stopwords

stop = stopwords.words('english')
without_stop_words = [word for word in without_capital_letters if word not in stop]

print (without_stop_words)
```

```python
import string
exclude = set(string.punctuation)
without_punct = [word for word in without_stop_words if word not in exclude]

print (without_punct)
```

```python
import pandas as pd
df = pd.DataFrame({'col':tweets_text})
df
```

```python
import codecs, nltk, string
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import gensim
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def nlp_pipeline(text):
    
    # if you want you can split in sentences - i'm usually skipping this step
    # text = nltk.sent_tokenize(text) 
    text=text.lower()
    
    #tokenize words for each sentence
    text = nltk.word_tokenize(text)
   
    # remove punctuation and numbers
    text = [token for token in text if token not in exclude and token.isalpha()]
    
    # remove stopwords - be careful with this step    
    text = [token for token in text if token not in without_stop_words]
    
    # pos tagger
    text = nltk.pos_tag(text)

    return text
```

```python
df["col"]=df["col"].astype(str) 
df.head()
```

```python
df['clean'] = df['col'].apply(nlp_pipeline)

df
```

```python
df.clean = df.clean.apply(lambda x: [(t[0],) for t in x if t[1]=='JJ' or t[1]=='JJR' or t[1]=='JJC'])
```

```python
df.head()
```

```python
df["clean"] = [','.join(map(str, l)) for l in df["clean"]]
```

```python
def stringsonly(text):
    text = text.replace(',', '')
    text = text.replace(')', '')
    text = text.replace('(', '')
    text = text.replace("''", ' ')
    text = text.replace("'", ' ')
    return text

df['clean'] = df['clean'].apply(stringsonly)
df
```

## Hashtags

```python
search2 = twitter_api.search(q='#trump',count=1000,result_type='recent')
tweets = search2['statuses']
```

```python
tweets_text = []

for tweet in tweets:
    text = tweet["text"]
    
    # add tweets to list
    tweets_text.append(text)
print (tweets_text)

```

```python
hashtags = []

for tweet in tweets_text:
    tweet = tweet.split(" ")
    for word in tweet:
        if "#" in word:
            hashtags.append(word)
            
```

```python
hashtags
```

```python
from collections import Counter
hashtags_to_count = (word for word in hashtags)
c = Counter(hashtags_to_count)
print (c.most_common(20))
```
